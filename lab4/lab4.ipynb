```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.array([1000000, 500000, 200000, 100000, 50000])
I_c = np.array([0.0098, 0.0195, 0.0478, 0.0931, 0.1805])*10**-3
I_b = (5-0.6)/(R+4.7*10**3)
```

```python
x = I_b
y=I_c

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
# poly1d_fn is now a function which takes in x and returns an estimate for y
plt.plot(x,y, 'yo')
plt.plot(x, poly1d_fn(x), '--k', label = "beta = " + str(coef[0]))
plt.xlabel("I_b (A)")
plt.ylabel("I_c (A)")
plt.title("I_b vs I_c")
plt.legend()
plt.savefig("4-4.png")

```
