## 2-4

f_3db measurement experimentally <br>
analytic function at high and low frequency

```python
data = pd.read_csv("lab2 - Sheet1.csv")
y = data[["phase shift(degree)", "amplitude(v)"]].to_numpy()
phase = y[:,0]
amplitude = y[:,1]
amplitude_log = np.log10(amplitude)
x = np.logspace(np.log10(0.1*1125.35),np.log10(10*1125.35,), num=20)

capacitor = 9.866*10**-9
resistor = 14.711 * 10 ** 3
tau = capacitor * resistor

# amplitude
relative_amp = 21/(np.sqrt(1+(2*np.pi*x*tau)**2)) # x = frequency
relative_amp_log = np.log10(relative_amp)

fig, ax = plt.subplots(2, figsize = (5, 10))

ax[0].scatter(x, amplitude_log, s=60, alpha=0.7, edgecolors="k", label = "experiment") # log-log scatter
ax[0].semilogx(x, relative_amp_log, color = "red", label = "theory")
ax[0].set_xscale("log")
ax[0].set_title("Low pass amplitude plot")
ax[0].set_xlabel("Frequency (Hz)", fontsize = 8)
ax[0].set_ylabel("V_in/V_out", fontsize = 8)
ax[0].legend()


ax[1].scatter(x, phase, s=60, alpha=0.7, edgecolors="k")
ax[1].set_xscale("log")
ax[1].set_title("Low pass phase plot")
ax[1].set_xlabel("Frequency (Hz)", fontsize = 8)
ax[1].set_ylabel("Phase (degree)", fontsize = 8)

plt.show()

#print((15.2/20.6)/(1/np.sqrt(2)))





```

## 2-5

calculation of Bode plot and phase plot?

```python
y = data[["phase shift", "amp"]].to_numpy()
phase = y[:,0]
amplitude = y[:,1]
amplitude_log = np.log10(amplitude)
fig, ax1 = plt.subplots(2, figsize = (5, 10))


# amplitude
amp_theory = (2*np.pi*x*tau*21)/(np.sqrt(1+(2*np.pi*x*tau)**2)) # x = frequency
amp_theory_log = np.log10(amp_theory)

ax1[0].scatter(x, amplitude_log, s=60, alpha=0.7, edgecolors="k", label = "experiment") # log-log scatter
ax1[0].semilogx(x, amp_theory_log, color = "red", label = "theory")
ax1[0].set_xscale("log")
ax1[0].set_title("High pass amplitude plot")
ax1[0].set_xlabel("Frequency (Hz)", fontsize = 8)
ax1[0].set_ylabel("V_in/ V_out", fontsize = 8)
ax1[0].legend()


ax1[1].scatter(x, phase, s=60, alpha=0.7, edgecolors="k")
ax1[1].set_xscale("log")
ax1[1].set_title("High pass phase plot")
ax1[1].set_xlabel("Frequency (Hz)", fontsize = 8)
ax1[1].set_ylabel("Phase (degree)", fontsize = 8)

plt.show()
```
